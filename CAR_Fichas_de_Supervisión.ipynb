```python
from pdf2image import convert_from_path
import easyocr
import numpy as np
from PIL import Image, ImageDraw
import spacy
```

```python
# Subir el archivo PDF
uploaded = files.upload()
pdf_file = list(uploaded.keys())[0]  # Obtener el nombre del archivo PDF subido
```

```python
images = convert_from_path(pdf_file)
```

```python
# Muestra la primera página del PDF
from IPython.display import display, Image
display(images[1])
```

```python
import easyocr  # Asegúrate de que easyocr está importado

```

```python
# Crea una instancia de easyocr.Reader
reader = easyocr.Reader(['es'])  # 'en' es para inglés, puedes agregar más idiomas si es necesario
```

```python
# Realiza OCR en la primera imagen
bounds = reader.readtext(np.array(images[0]), min_size=0, slope_ths=0.2, ycenter_ths=0.7, height_ths=0.6, width_ths=0.8, decoder='beamsearch', beamWidth=10)
```

```python
 #Dibuja los cuadros alrededor del texto identificado
def draw_boxes(image, bounds, color='blue', width=4):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(images[0], bounds)
```

```python
# Extrae y muestra el texto
text = ''
for i in range(len(bounds)):
    text += bounds[i][1] + '\n'

print(text)
```
